```python
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('/content/Sample - Superstore.xls')
```

```python
df
```

```python
df.dtypes
```

# **Bar Graph**

```python
plt.figure(figsize=(13, 6))
plt.bar(df['State'],df['Sales'])
plt.ylim(0,20000)
plt.xticks(rotation=90)
plt.title('State vs Sales')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```


# **Histogram**




```python
plt.figure(figsize=(10,5))
plt.hist(df['Quantity'],color='red',rwidth=0.9)
plt.title('Distribution of Quantity',fontsize=10)
plt.xlabel('Quantity')
plt.ylabel('Distribution of Quantity')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **SCATTER PLOT**

```python
plt.figure(figsize=(8,8))
plt.scatter(df['Discount'],df['Profit'],c='violet',s=5,marker='*')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# **Heat Map**

```python
plt.figure(figsize=(8, 6))
df1=df[['Profit','Discount','Sales']]
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
